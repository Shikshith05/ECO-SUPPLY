# epoch — Exploratory Data Analysis

Use this notebook to explore your dataset before training models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
df = pd.read_csv('../data/raw/supply_chain.csv', encoding='utf-8', on_bad_lines='skip')
print(f'Shape: {df.shape}')
df.head()
```

```python
df.info()
```

```python
nulls = df.isnull().sum()
nulls[nulls > 0].sort_values(ascending=False)
```

```python
region_sales = df.groupby('Order Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', title='Total Sales by Region', color='steelblue')
plt.tight_layout(); plt.show()
```

```python
df['Late_delivery_risk'].value_counts().plot(kind='bar', title='Late Delivery Risk Distribution')
plt.xticks([0,1], ['On Time','Late'], rotation=0)
plt.tight_layout(); plt.show()
```

```python
pd.crosstab(df['Shipping Mode'], df['Late_delivery_risk'], normalize='index').plot(
    kind='bar', stacked=True, title='Shipping Mode vs Late Delivery Risk'
)
plt.tight_layout(); plt.show()
```
